```python
def custom_speaker_selection_func(last_speaker: Agent, groupchat: autogen.GroupChat):
    """Define a customized speaker selection function.
    A recommended way is to define a transition for each speaker in the groupchat.
    # User_console_and_code_interpreter,
    financial_news_research_analyst,
    central_banks_insights_specialist,
    global_monetary_policy_analyst,
    financial_data_analyst,
    web_surfer'

    Returns:
        Return an `Agent` class or a string from ['auto', 'manual', 'random', 'round_robin'] to select a default method to use.
    """
    messages = groupchat.messages
    list_of_speakers = [m['name'] for m in messages]

    if (
        last_speaker != agents_dict['financial_news_research_analyst']
        or "TERMINATE" not in messages[-1]["content"]
    ):
        # default to auto speaker selection method
        return "auto"
    if "central_banks_insights_specialist" in list_of_speakers:
        return agents_dict["central_banks_insights_specialist"]
    elif "global_monetary_policy_analyst" in list_of_speakers:
        return agents_dict["global_monetary_policy_analyst"]

```

```python
def start_task(execution_task: str, agents: Dict[str, Union[str, List[str]]]):
   group_chat = autogen.GroupChat(agents=list(agents.values()),  # type: ignore
                                   messages=[], 
                                   max_round=30,
                                   speaker_selection_method=custom_speaker_selection_func,
                                   allow_repeat_speaker=False,
                                   )
    
   manager = autogen.GroupChatManager(groupchat=group_chat, 
                                       llm_config={"config_list": config_list, **llm_config},
                                       human_input_mode="TERMINATE",
                                       )
   
   agent_list[0].initiate_chat(manager, message=execution_task, cache=None)
```
